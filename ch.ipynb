```python
def sum_n(n):
    
    if n <= 0:
        return 0
    return n + sum_n(n-1)

if __name__ == '__main__':
    try:
        n = int(input().strip())
    except Exception:
        n = 0
    print(sum_n(n))
```

```python

def factorial(n):
    
    
    if n <= 1:
        return 1
    return n * factorial(n-1)

if __name__ == '__main__':
    try:
        n = int(input().strip())
    except Exception:
        n = 0
    print(factorial(n))

```

```python
def print_1_to_n(n):
    
    def helper(i):
        if i > n:
            return
        if i == n:
            print(i, end='')
        else:
            print(i, end=' ')
        helper(i+1)
    if n >= 1:
        helper(1)

if __name__ == '__main__':
    try:
        n = int(input().strip())
    except Exception:
        n = 0
    if n > 0:
        print_1_to_n(n)
    print()
```

```python
def print_n_to_1(n):
    
    def helper(i):
        if i < 1:
            return
        if i == 1:
            print(i, end='')
        else:
            print(i, end=' ')
        helper(i-1)
    if n >= 1:
        helper(n)

if __name__ == '__main__':
    try:
        n = int(input().strip())
    except Exception:
        n = 0
    if n > 0:
        print_n_to_1(n)
    print()
```

```python
def fibonacci(n):
    
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

if __name__ == '__main__':
    try:
        n = int(input().strip())
    except Exception:
        n = 0
    print(fibonacci(n))
```

```python
def power(x, n):
    """Return x raised to the power n using recursion."""
    if n == 0:
        return 1
    if n < 0:
        return 1 / power(x, -n)
    return x * power(x, n-1)

if __name__ == '__main__':
    try:
        parts = input().strip().split()
        if len(parts) >= 2:
            x = int(parts[0])
            n = int(parts[1])
        else:
            x = 0
            n = 0
    except Exception:
        x = 0
        n = 0
    res = power(x, n)
    # Print as int when result is integral
    if isinstance(res, float) and res.is_integer():
        print(int(res))
    else:
        print(res)
```

```python
def count_digits(n):
    """Return the number of digits in integer n using recursion."""
    n = abs(n)
    if n == 0:
        return 1
    if n < 10:
        return 1
    return 1 + count_digits(n // 10)

if __name__ == '__main__':
    try:
        n = int(input().strip())
    except Exception:
        n = 0
    print(count_digits(n))
```
